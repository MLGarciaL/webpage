```python
import pandas as pd
import numpy as np

def mean_exits_after_cutoff(sim, cutoff=25):
    exit_records = []

    # Collect exits from all industries
    for ind_id, ind in enumerate(sim.industries):
        for d in ind.dead_firms:
            exit_records.append({
                'industry': ind_id,
                'enter_time': d['enter_time'],
                'exit_time': d['exit_time']
            })

    df_exit = pd.DataFrame(exit_records)

    if df_exit.empty:
        return pd.DataFrame(columns=['exit_time', 'mean_exits', 'cum_mean_exits'])

    # Filter survivors until the cutoff
    df_exit_filtered = df_exit[
        (df_exit['enter_time'] < cutoff) &
        (df_exit['exit_time'] > cutoff)
    ][['industry','exit_time']]

    if df_exit_filtered.empty:
        return pd.DataFrame(columns=['exit_time', 'mean_exits', 'cum_mean_exits'])

    # Count exits by industry + exit_time
    exit_counts = (
        df_exit_filtered
        .groupby(['industry', 'exit_time'])
        .size()
        .reset_index(name='num_exits')
    )

    # Mean exits per exit_time across industries
    mean_exits_by_time = (
        exit_counts
        .groupby('exit_time')['num_exits']
        .mean()
        .reset_index(name='mean_exits')
    )

    # Fill missing times
    full_times = pd.DataFrame({'exit_time': range(cutoff - 1 , cutoff*2 + 1)})
    mean_exits_by_time = full_times.merge(mean_exits_by_time, on='exit_time', how='left')
    mean_exits_by_time['mean_exits'] = mean_exits_by_time['mean_exits'].fillna(0)

    # Cumulative mean exits with forward fill
    mean_exits_by_time['cum_mean_exits'] = mean_exits_by_time['mean_exits'].cumsum()

    return mean_exits_by_time
```

```python
results410 = mean_exits_after_cutoff(sim410)
results411 = mean_exits_after_cutoff(sim411)
results415 = mean_exits_after_cutoff(sim415)

import matplotlib.pyplot as plt

plt.plot(results410['exit_time'], (100 - results410['cum_mean_exits'])/(100 - results410['cum_mean_exits'])*100, linestyle='-', lw = 2, color = 'black', label = 'K = 0')
plt.plot(results411['exit_time'], (100 - results411['cum_mean_exits'])/(100 - results410['cum_mean_exits'])*100, linestyle='--', lw = 2, color = 'black', label = 'K = 1')
plt.plot(results415['exit_time'], (100 - results415['cum_mean_exits'])/(100 - results410['cum_mean_exits'])*100, linestyle=':', lw = 2, color = 'black', label = 'K = 5')
plt.xlabel('Time')
plt.title('% Surviving Incumbents', loc = 'left')
plt.ylabel('Cumulative Mean Exits')
plt.suptitle("Figure 6 Survival in Changing Environments \n (Change in Fitness Contribution of One Dimension)", x = .44, y = 1.05, fontweight='bold', fontsize=12)
plt.figtext(0.1, 0.01, "Replicated from Levinthal (1997)", ha="left", fontsize=9, style='italic')
plt.savefig('Figure6.png', dpi=300, bbox_inches='tight')
plt.ylim(0,110)
plt.legend()
plt.show()
```

```python
import pandas as pd

def long_jump_frequency_over_time(sim, cutoff=25):
    """
    Calculate the frequency of at least one long jump for firms that survived vs. 
    those that died, for each time period after a given cutoff.

    Parameters:
        sim : IndustrySimulation
            The simulation object containing industries and firms.
        cutoff : int
            Time cutoff; only firms that entered before this time are considered.
    
    Returns:
        pd.DataFrame with columns:
            'time' : time period
            'survival' : 'survived' or 'died'
            'num_firms' : number of firms in this category at that time
            'num_with_long_jump' : number of firms with at least one long jump before that time
            'frequency' : fraction of firms with at least one long jump
    """
    
    records = []

    # Determine the maximum time across industries
    max_time = max(ind.history_industry[-1]['time'] for ind in sim.industries if ind.history_industry)

    for t in range(cutoff, max_time + 1):
        for ind_id, ind in enumerate(sim.industries):
            all_firms = ind.firms + [d['firm'] for d in ind.dead_firms]
            for f in all_firms:
                if f.timeborn >= cutoff:
                    continue  # skip firms entering after cutoff

                # Did the firm survive beyond time t?
                exited = next((d['exit_time'] for d in ind.dead_firms if d['firm'] is f), None)
                survived_at_t = exited is None or exited > t
                survival_status = 'survived' if survived_at_t else 'died'

                # Did the firm perform at least one long jump before time t?
                did_long_jump = any(
                    h['type of search'] == 2 and h.get('time', 0) <= t
                    for h in f.historial
                )

                records.append({
                    'time': t,
                    'industry': ind_id,
                    'firm': f,
                    'survival': survival_status,
                    'long_jump': did_long_jump
                })

    df = pd.DataFrame(records)

    # Aggregate by time and survival status
    summary = (
        df.groupby(['time','survival'])['long_jump']
        .agg(['sum','count'])
        .reset_index()
        .rename(columns={'sum':'num_with_long_jump', 'count':'num_firms'})
    )
    summary['frequency'] = summary['num_with_long_jump'] / summary['num_firms']

    return summary
```

```python
freq_time = long_jump_frequency_over_time(sim415, cutoff=25)
print(freq_time.head())
```

```python
# Pivot the table to have survivors and died in separate columns
pivot_freq = freq_time.pivot(index='time', columns='survival', values='frequency').reset_index()

# Compute the ratio: survivors / dead
pivot_freq['ratio_survived_dead'] = pivot_freq['survived'] / pivot_freq['died']

# Pivot the table to have survivors and died in separate columns
pivot_freq1 = freq_time1.pivot(index='time', columns='survival', values='frequency').reset_index()

# Compute the ratio: survivors / dead
pivot_freq1['ratio_survived_dead'] = pivot_freq1['survived'] / pivot_freq1['died']
```
